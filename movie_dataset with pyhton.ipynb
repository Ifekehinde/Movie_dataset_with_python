```python
import pandas as pd
```

```python
df = pd.read_csv('assignment.csv')
```

```python
df.head()
```

```python
df.info()
```

### 1. Convert columns header to lower case

```python
df.columns = df.columns.str.lower()
```

```python
df
```

### 2. Get rid of weird characters in columns header

```python
df.columns = df.columns.str.replace('original titlê','original title')
```

```python
df
```

```python
df.columns = df.columns.str.replace('genrë¨','genre')
```

```python
df
```

### 3. Get rid of unnamed columns

```python
df.pop('unnamed: 0')
df.pop('unnamed: 8')
```

```python
df
```

### 4. Replace spaces between columns header with _

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df
```

```python
df.columns = df.columns.str.replace ('_votes_', 'votes')
```

```python
df
```

### 5. format date column to 'YMD'

```python
from datetime import datetime
```

```python
df['release_year'] = pd.to_datetime(df['release_year'], errors='coerce')
```

```python
df['release_year'] = df['release_year'].dt.strftime('%Y-%m-%d')
```

```python
df
```

### 6. Return only the first genre in the genre column

```python
df.genre.str.split(',', expand= True)
```

```python
df.genre = df.genre.str.split(',', expand= True)[0]
```

```python
df
```

### 7. Format country column to unique country names

```python
df.country.value_counts()
```

```python
df.country = df.country.str.replace('US.','USA', regex = True)
df.country = df.country.str.replace('USAA','USA', regex = True)
df.country = df.country.str.replace('West Germany', 'Germany', regex = True)
df.country = df.country.str.replace('New Zealand', 'New Zeland', regex = True)
df.country = df.country.str.replace('New Zesland', 'New Zeland', regex = True)
df.country = df.country.str.replace('Italy1', 'Italy', regex = True)
```

```python
df.country.value_counts()
```

### 8. Convert income column to an integer and get rid of dollar sign

```python
df.income = df.income.str.replace('$','', regex = True)
df.income = df.income.str.replace(',','', regex = True)
df.income = df.income.str.replace('o','0', regex = True)
```

```python
df
```

```python
df.income = pd.to_numeric(df.income, errors='coerce')
df.income = df.income.astype(pd.Int64Dtype())
```

```python
df.info()
```

### 9. Convert votes column to an integer

```python
df.votes = df.votes.str.replace('.', '', regex = True)
```

```python
df
```

```python
df.votes.value_counts()
```
